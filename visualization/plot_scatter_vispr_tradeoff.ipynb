```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as patches


pap_lowres = [68.13, 57.64, 51.99, 47.18, 43.72, 41.21]
action_lowres = [67.90, 50.22, 43.42, 32.8, 28.95, 26.16]
NR_action = [67.36, 66.27, 66.89, 66.83]
NR_pap = [66.68, 66.94, 65.59, 65.12]
R_action = [67.58, 65.46, 66.28, 67.59]
R_pap = [65.28, 64.82, 63.58, 62.06]
```

```python
fig = plt.figure(figsize=(20, 10))
ax = fig.gca()
plt.scatter(pap_lowres, action_lowres, s=0.2*np.arange(1,11)**4 , c='r', marker='o',label='Method 1')
plt.scatter(NR_pap, NR_action, s=40*np.arange(1,5)**2, c='g', marker=(5, 1), label='Method 4')
plt.scatter(R_pap, R_action, s=40*np.arange(1,5)**2, c='b', marker=(5, 1), label='Method 5')
plt.grid(True)
plt.plot([40, 70], [40, 70], ls="--", c='k')
plt.axes().set_aspect('equal')
plt.axis([40, 70, 40, 70])
lgnd=plt.legend( prop={'size': 24}, loc=2, markerscale=1.0)
lgnd.legendHandles[0]._sizes = [240]
lgnd.legendHandles[1]._sizes = [240]
lgnd.legendHandles[2]._sizes = [240]
#rect = plt.Rectangle((30, 60), 20, 20, color='r', ls="--", alpha=1, linewidth=2,fill=False)
#ax.add_patch(rect)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.ylabel('Action Accuracy', fontsize=15)
plt.xlabel('Privacy Attributes Prediction cMAP', fontsize=15)
#plt.title('Action & Privacy Attributes Prediction Performance Tradeoff',fontsize=20)
plt.savefig('vispr_tradeoff_17attributes.eps', bbox_inches='tight',transparent=True, pad_inches=0)
plt.show()
```
