```python
def plot_bar(models, R2, NR2, R4, NR4, R8, NR8, R12, NR12):
    import operator
    # this is for plotting purpose
    plt.figure(figsize=(80,45))
    index = np.arange(len(models))
    plt.bar(index, R2, width=0.2, color='r', edgecolor='black', hatch="x", align='center', label='R2')
    plt.bar(index, list(map(operator.sub, NR2, R2)), bottom=R2, width=0.2, color='b', edgecolor='black', hatch="x", align='center', label='NR2')

    plt.bar(index+0.2, R4, width=0.2, color='r', edgecolor='black', hatch='*' ,align='center', label='R4')
    plt.bar(index+0.2, list(map(operator.sub, NR4, R4)), bottom=R4, width=0.2, color='b', edgecolor='black', hatch="*",align='center', label='NR4')
    
    plt.bar(index+0.4, R8, width=0.2, color='r', edgecolor='black', hatch="O", align='center', label='R8')
    plt.bar(index+0.4, list(map(operator.sub, NR8, R8)), bottom=R8, width=0.2, color='b', edgecolor='black', hatch="O", align='center', label='NR8')

    plt.bar(index+0.6, R12, width=0.2, color='r', edgecolor='black', hatch=".", align='center', label='R12')
    plt.bar(index+0.6, list(map(operator.sub, NR12, R12)), bottom=R12, width=0.2, color='b', edgecolor='black', hatch=".", align='center', label='NR12')
    
    plt.xlabel('Model Name', fontsize=60)
    plt.ylabel('Actor Pair Recognition Accuracy', fontsize=60)
    plt.xticks(index+0.2, models, fontsize=60, rotation=75)
    plt.yticks(fontsize=60)
    plt.legend( prop={'size': 80})
    plt.title('Actor Pair Recognition using N=10 Evaluation Models',fontsize=90)
    plt.savefig('sbu_actor_acc.png')
    plt.show()
```

```python
plot_bar(models, R2, NR2, R4, NR4, R8, NR8, R12, NR12)
```
